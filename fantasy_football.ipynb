### **تحلیل تک متغیره**

```python
# تابع رسم نمودار جعبه ای و هیستوگرام در امتداد یک مقیاس


def histogram_boxplot(data, feature, figsize = (12, 7), kde = False, bins = None):
    
    """
    نمودار جعبه ای و هیستوگرام ترکیب شده اند

    data: داده
    feature: ویژگی داده
    figsize: اندازه شکل (پیش فرض (7، 12))
    kde: آیا منحنی تراکم نشان داده شود یا خیر (به صورت پیش فرض خیر)
    bins: تعداد سطل های هیستوگرام (پیش فرض هیچ)
    """
    
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,  # تعداد ردیف های شبکه زیرنمودار = 2
        sharex = True,  # در میان همه زیرنمودارها به اشتراک گذاشته می شود X محور
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize,
    )  # ایجاد دو زیرنمودار
    sns.boxplot(
        data = data, x = feature, ax = ax_box2, showmeans = True, color = "violet"
    )  # یک نمودار جعبه ای ایجاد خواهد شد و یک ستاره نشان دهنده مقدار میانگین ستون خواهد بود
    sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, palette = "winter"
    ) if bins else sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2
    )  # برای نمودار هیستوگرام
    ax_hist2.axvline(
        data[feature].mean(), color = "green", linestyle = "--"
    )  # اضافه کردن میانگین به نمودار هیستوگرام
    ax_hist2.axvline(
        data[feature].median(), color = "black", linestyle = "-"
    )  # اضافه کردن میانه به نمودار هیستوگرام
```

```python
# تابع برای ایجاد نمودارهای میله ای برچسب دار


def labeled_barplot(data, feature, perc = False, n = None):
    """
    نمودار میله ای با نشان دادن درصد در بالای آن

    data: داده
    feature: ویژگی داده
    perc: آیا باید درصد به جای تعداد نشان داده شوند ؟ (به صورت پیش فرض خیر)
    n:سطح برتر دسته بندی را نشان میدهد(پیشفرض هیج است، به این معنی که تمام سطوح نمایش داده می شود) n
    """

    total = len(data[feature])  # طول ستون
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize = (count + 1, 5))
    else:
        plt.figure(figsize = (n + 1, 5))

    plt.xticks(rotation = 90, fontsize = 15)
    ax = sns.countplot(
        data = data,
        x = feature,
        palette = "Paired",
        order = data[feature].value_counts().index[:n].sort_values(),
    )

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # درصد هر کلاس از دسته بندی
        else:
            label = p.get_height()  # تعداد هر سطح از دسته بندی

        x = p.get_x() + p.get_width() / 2  # عرض نمودار
        y = p.get_height()  # طول نمودار

        ax.annotate(
            label,
            (x, y),
            ha = "center",
            va = "center",
            size = 12,
            xytext = (0, 5),
            textcoords = "offset points",
        )  # نوشتن درصد

    plt.show()  # نمایش نمودار
```

```python
k_means_df = data_pca.copy()
```

```python
clusters = range(1, 15)
meanDistortions = []

for k in clusters:
    model = KMeans(n_clusters = k, random_state = 1)
    model.fit(data_pca)
    prediction = model.predict(k_means_df)
    distortion = (
        sum(np.min(cdist(k_means_df, model.cluster_centers_, "euclidean"), axis = 1))
        / k_means_df.shape[0]
    )

    meanDistortions.append(distortion)

    print("Number of Clusters:", k, "\tAverage Distortion:", distortion)

plt.plot(clusters, meanDistortions, "bx-")
plt.xlabel("k")
plt.ylabel("Average Distortion")
plt.title("Selecting k with the Elbow Method", fontsize = 20)
plt.show()
```

# خوشه بندی سلسله مراتبی

```python
hc_df = data_pca.copy()
```

```python
# فهرست معیارهای فاصله
distance_metrics = ["euclidean", "chebyshev", "mahalanobis", "cityblock"]

# لیست روش های پیوند
linkage_methods = ["single", "complete", "average", "weighted"]

high_cophenet_corr = 0
high_dm_lm = [0, 0]

for dm in distance_metrics:
    for lm in linkage_methods:
        Z = linkage(hc_df, metric = dm, method = lm)
        c, coph_dists = cophenet(Z, pdist(hc_df))
        print(
            "Cophenetic correlation for {} distance and {} linkage is {}.".format(
                dm.capitalize(), lm, c
            )
        )
        if high_cophenet_corr < c:
            high_cophenet_corr = c
            high_dm_lm[0] = dm
            high_dm_lm[1] = lm
            
# cophenetic چاپ ترکیبی از فاصله و روش پیوند با بالاترین همبستگی
print('*'*100)
print(
    "Highest cophenetic correlation is {}, which is obtained with {} distance and {} linkage.".format(
        high_cophenet_corr, high_dm_lm[0].capitalize(), high_dm_lm[1]
    )
)
```

**بیایید روش های مختلف پیوند را فقط با فاصله اقلیدسی بررسی کنیم.**

```python
# لیست روش های پیوند
linkage_methods = ["single", "complete", "average", "centroid", "ward", "weighted"]

high_cophenet_corr = 0
high_dm_lm = [0, 0]

for lm in linkage_methods:
    Z = linkage(hc_df, metric = "euclidean", method = lm)
    c, coph_dists = cophenet(Z, pdist(hc_df))
    print("Cophenetic correlation for {} linkage is {}.".format(lm, c))
    if high_cophenet_corr < c:
        high_cophenet_corr = c
        high_dm_lm[0] = "euclidean"
        high_dm_lm[1] = lm
        
# cophenetic چاپ ترکیبی از فاصله و روش پیوند با بالاترین همبستگی
print('*'*100)
print(
    "Highest cophenetic correlation is {}, which is obtained with {} linkage.".format(
        high_cophenet_corr, high_dm_lm[1]
    )
)
```

**حال دندروگرام ها را برای روش های مختلف پیوند مشاهده می کنیم.**

```python
# لیست روش های پیوند
linkage_methods = ["single", "complete", "average", "centroid", "ward", "weighted"]

# cophenetic لیست هایی برای ذخیره نتایج محاسبه همبستگی 
compare_cols = ["Linkage", "Cophenetic Coefficient"]
compare = []

# subplot برای ایجاد یک تصویر 
fig, axs = plt.subplots(len(linkage_methods), 1, figsize = (15, 30))

#را محاسبه می کنیم cophenetic برای هر روش پیوند، دندروگرام را رسم می کنیم و همبستگی
for i, method in enumerate(linkage_methods):
    Z = linkage(hc_df, metric = "euclidean", method = method)

    dendrogram(Z, ax = axs[i])
    axs[i].set_title(f"Dendrogram ({method.capitalize()} Linkage)")

    coph_corr, coph_dist = cophenet(Z, pdist(hc_df))
    axs[i].annotate(
        f"Cophenetic\nCorrelation\n{coph_corr:0.2f}",
        (0.80, 0.80),
        xycoords="axes fraction",
    )

    compare.append([method, coph_corr])
```

## **DBSCAN**

DBSCAN یک الگوریتم بسیار قدرتمند برای یافتن خوشه‌های با چگالی بالا است، اما مشکل آن ، تعیین بهترین مجموعه ابرپارامترها برای استفاده است که شامل دو نمونه ابرپارامتر eps و min است.

از آنجایی که DBSCAN یک الگوریتم بدون نظارت است، برخلاف الگوریتم یادگیری با نظارت ، که به شما امکان می دهد الگوریتم خود را بر روی یک مجموعه تست آزمایش کنید، این امکان در اینجا وجود ندارد. رویکردی که می‌توانیم دنبال کنیم اساساً آزمایش مجموعه‌ای از ترکیب‌های مختلف مقادیر و یافتن امتیاز silhouette برای هر یک از آنهاست.

### **نمره silhouette چه چیزی  است ؟**

امتیاز Silhouette یکی از روش‌های ارزیابی کیفیت خوشه‌های ایجاد شده با استفاده از الگوریتم‌های خوشه‌بندی مانند K-Means است. امتیاز silhouette معیاری است که نشان می دهد یک شی چقدر شبیه به خوشه خود و چقدر متفاوت با سایر خوشه ها است. امتیاز Silhouette دارای محدوده [1، 1-] است.

* ضرایب Silhouette نزدیک به 1+ نشان می دهد که نمونه از خوشه های همسایه دور است. 
* امتیاز Silhouette نزدیک به 1- نشان می دهد که آن نمونه ها ممکن است به خوشه اشتباهی اختصاص داده شده باشند.

```python
dbscan_df = data_pca.copy()
```

```python
# مقداردهی اولیه لیست ها
eps_value = [2,3]  #رندوم در نظر می گیریم eps یک
min_sample_values = [6,20]  #رندوم در نظر می گیریم min_sample یک

# ایجاد یک دیکشنری با توجه به مقادیر بالا
res = {eps_value[i]: min_sample_values for i in range(len(eps_value))}  
```

```python
#برای هر یک از ترکیب ها silhouette_score پیدا کردن    

high_silhouette_avg = 0                                               
high_i_j = [0, 0]                                                   
key = res.keys()                                                      
for i in key:
    z = res[i]                                                       
    for j in z:
        db = DBSCAN(eps = i, min_samples = j).fit(dbscan_df)               # استفاده از الگوریتم برای هر یک از ترکیبات موجود در دیکشنری
        core_samples_mask = np.zeros_like(db.labels_, dtype = bool)
        core_samples_mask[db.core_sample_indices_] = True
        labels = db.labels_
        silhouette_avg = silhouette_score(dbscan_df, labels)           #silhouette پیدا کردن امتیاز
        print( 
            "For eps value =" + str(i),
            "For min sample =" + str(j),
            "The average silhoutte_score is :",
            silhouette_avg,                                          #برای هر یک از ترکیب ها silhouette چاپ امتیاز
        )
        if high_silhouette_avg < silhouette_avg:                                   
            high_i_j[0] = i
            high_i_j[1] = j
```

```python
# Printing the highest silhouette score
print(
    "Highest_silhoutte_avg is {} for eps = {} and min sample = {}".format(
        high_silhouette_avg, high_i_j[0], high_i_j[1]
    )
)
```
