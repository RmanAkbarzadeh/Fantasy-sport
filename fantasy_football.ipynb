## **توضیحات مجموعه داده**
--------------------------

- **Player_Name:** اسم بازیکن.

### **تحلیل تک متغیره**

```python
# تابع رسم نمودار جعبه ای و هیستوگرام در امتداد یک مقیاس


def histogram_boxplot(data, feature, figsize = (12, 7), kde = False, bins = None):
    
    """
    نمودار جعبه ای و هیستوگرام ترکیب شده اند

    data: داده
    feature: ویژگی داده
    figsize: اندازه شکل (پیش فرض (7، 12))
    kde: آیا منحنی تراکم نشان داده شود یا خیر (به صورت پیش فرض خیر)
    bins: تعداد سطل های هیستوگرام (پیش فرض هیچ)
    """
    
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,  # تعداد ردیف های شبکه زیرنمودار = 2
        sharex = True,  # در میان همه زیرنمودارها به اشتراک گذاشته می شود X محور
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize,
    )  # ایجاد دو زیرنمودار
    sns.boxplot(
        data = data, x = feature, ax = ax_box2, showmeans = True, color = "violet"
    )  # یک نمودار جعبه ای ایجاد خواهد شد و یک ستاره نشان دهنده مقدار میانگین ستون خواهد بود
    sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, palette = "winter"
    ) if bins else sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2
    )  # برای نمودار هیستوگرام
    ax_hist2.axvline(
        data[feature].mean(), color = "green", linestyle = "--"
    )  # اضافه کردن میانگین به نمودار هیستوگرام
    ax_hist2.axvline(
        data[feature].median(), color = "black", linestyle = "-"
    )  # اضافه کردن میانه به نمودار هیستوگرام
```

```python
# تابع برای ایجاد نمودارهای میله ای برچسب دار


def labeled_barplot(data, feature, perc = False, n = None):
    """
    نمودار میله ای با نشان دادن درصد در بالای آن

    data: داده
    feature: ویژگی داده
    perc: آیا باید درصد به جای تعداد نشان داده شوند ؟ (به صورت پیش فرض خیر)
    n:سطح برتر دسته بندی را نشان میدهد(پیشفرض هیج است، به این معنی که تمام سطوح نمایش داده می شود) n
    """

    total = len(data[feature])  # طول ستون
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize = (count + 1, 5))
    else:
        plt.figure(figsize = (n + 1, 5))

    plt.xticks(rotation = 90, fontsize = 15)
    ax = sns.countplot(
        data = data,
        x = feature,
        palette = "Paired",
        order = data[feature].value_counts().index[:n].sort_values(),
    )

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # درصد هر کلاس از دسته بندی
        else:
            label = p.get_height()  # تعداد هر سطح از دسته بندی

        x = p.get_x() + p.get_width() / 2  # عرض نمودار
        y = p.get_height()  # طول نمودار

        ax.annotate(
            label,
            (x, y),
            ha = "center",
            va = "center",
            size = 12,
            xytext = (0, 5),
            textcoords = "offset points",
        )  # نوشتن درصد

    plt.show()  # نمایش نمودار
```

### **مقیاس بندی**

- قبل از اینکه به خوشه بندی ادامه دهیم، داده ها را مقیاس بندی می کنیم.

```python
# مقیاس بندی داده ها قبل از خوشه بندی
scaler = StandardScaler()
subset = df.iloc[:, 3:].copy()
subset_scaled = scaler.fit_transform(subset)
```

```python
# ایجاد یک دیتافریم از داده های مقیاس شده
subset_scaled_df = pd.DataFrame(subset_scaled, columns = subset.columns)
```

### **استفاده از PCA**

```python
# PCA وارد کردن 
from sklearn.decomposition import PCA

# تعریف تعداد اجزای اصلی برای تولید
n = subset.shape[1]                                         # ذخیره تعداد متغیرها در داده ها

pca = PCA(n_components = n, random_state = 1)                       # PCA ذخیره عملکرد

data_pca = pd.DataFrame(pca.fit_transform(subset_scaled_df ))          # روی داده های مقیاس شده PCA استفاده از 

exp_var = (pca.explained_variance_ratio_)                       
```

## **خوشه بندی K-Means**

```python
k_means_df = data_pca.copy()
```

```python
clusters = range(1, 15)
meanDistortions = []

for k in clusters:
    model = KMeans(n_clusters = k, random_state = 1)
    model.fit(data_pca)
    prediction = model.predict(k_means_df)
    distortion = (
        sum(np.min(cdist(k_means_df, model.cluster_centers_, "euclidean"), axis = 1))
        / k_means_df.shape[0]
    )

    meanDistortions.append(distortion)

    print("Number of Clusters:", k, "\tAverage Distortion:", distortion)

plt.plot(clusters, meanDistortions, "bx-")
plt.xlabel("k")
plt.ylabel("Average Distortion")
plt.title("Selecting k with the Elbow Method", fontsize = 20)
plt.show()
```

**مشاهدات:**

- از نقطه 4، نمودار شروع به حرکت تقریباً موازی با محور X می کند. مقدار K مربوط به این نقطه، مقدار K بهینه یا تعداد بهینه خوشه ها است.

**با k = 4 جلو می رویم.**

```python
kmeans = KMeans(n_clusters = 4, random_state = 1)
kmeans.fit(k_means_df)
```

```python
# ایجاد یک کپی از داده های اصلی
df1 = df.copy()

# به دیتافریم های اصلی K-Means اضافه کردن برچسب های خوشه بندی
k_means_df["KM_segments"] = kmeans.labels_
df1["KM_segments"] = kmeans.labels_
```
